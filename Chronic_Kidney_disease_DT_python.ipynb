```
# Numlist=[Weight,Height,BMI,Obese,Waist,SBP,DBP,HDL,LDL,Total Chol]
Objlist=['Female'              ,
'Educ'                ,
'Unmarried'           ,
'Income'              ,
'CareSource'          ,
'Insured'             ,
'Obese'               ,
'Dyslipidemia'        ,
'PVD',
'Activity'         ,
'PoorVision'          ,
'Smoker'              ,
'Hypertension'        ,
# 'Fam Hypertension'    ,
'Diabetes'            ,
# 'Fam Diabetes'        ,
'Stroke'              ,
'CVD'                 ,

'CHF'                 ,
'Anemia'                     ]


for i in Objlist:
  df[i]=df[i].astype(object)
```

```
df['Fam Diabetes']=df['Fam Diabetes'].astype('object')
df['Fam Diabetes']=df['Fam Diabetes'].astype('object')
```

```
df['Fam CVD']=df['Fam CVD'].astype('object')
```

```
df.dtypes
```

```
#imputing numeric variables with mean values
df=df.fillna(df.mean())
```

```
##Converting all Numeric variables as int
numlist=['Height',
'Weight',
'BMI',
'Waist',
'SBP',
# 'DBPHDL',
'LDL',
'Total Chol']

for i in numlist:
  df[i]=df[i].astype(int)
```

```
df['DBP']=df['DBP'].astype(int)
df['HDL']=df['HDL'].astype(int)
```

```
#Target Variable
df['CKD']=df['CKD'].astype('object')
```

```
##train test 
y_train=df['CKD']
x_train=df.drop(['CKD'],axis=1)
```

```
##train test split
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(x_train,y_train,test_size=0.2)
```
